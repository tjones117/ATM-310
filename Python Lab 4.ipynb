```python
import csv
import math
import decimal
import numpy as np
from matplotlib import pyplot as plt
```

```python
pzero = 1013 #hPa
tzero = 293 #K
rho = 1.225 #kg/m^3
g = 9.8 #m/s^2
Rd = 287
tau = 6.5

def compute_P(z):
    p = pzero - ((rho * g * z) / 100)
    return p

def compute_T(z):
    H = (Rd * tzero) / g
    p = pzero * math.exp(-z / H) 
    return p

def compute_tau(t):
    E = g / (Rd * tau)
    p = (pzero * (((t + 273) / tzero)**E)) * 100
    return p
```

```python
t = []
p = []
cP = []
cT = []
ctau = []

with open('sounding.txt', 'r') as f:
    for line in f.readlines():
        split_lines = line.split()
        p.append(float(split_lines[0]))
        t.append(float(split_lines[2]))
        cP.append(compute_P(float(split_lines[1])))
        cT.append(compute_T(float(split_lines[1])))
        ctau.append(compute_tau(float(split_lines[2])))
```

```python
fig = plt.figure()
plt.plot(t, p, color = 'r')
plt.plot(t, cP, color = 'b')
plt.title('P vs T (Constant rho)')
plt.ylabel('Pressure (hPa)')
plt.xlabel('Temperature (K)')

#fig.savefig('P vs T constant rho.pdf')
```

```python
fig = plt.figure()
plt.plot(t, p, color = 'r')
plt.plot(t, cT, color = 'b')
plt.title('P vs T (Constant T)')
plt.ylabel('Pressure (hPa)')
plt.xlabel('Temperature (K)')

#fig.savefig('P vs T constant T.pdf')
```

```python
fig = plt.figure()
plt.plot(t, p, color = 'r')
plt.plot(t, ctau, color = 'b')
plt.title('P vs T (Constant tau)')
plt.ylabel('Pressure (hPa)')
plt.xlabel('Temperature (K)')
```
